# MAP Visualisation

```r
#cleaning dataset 
honeypot_map <- honeypot %>% select(1:8, 10, 11,14,15,16)
dim(honeypot_map)
```

```r
cleaned_geolocations <- data.frame(honeypot_map %>% filter(!is.na(latitude) & !is.na(longitude)))
dim(cleaned_geolocations)
```

```r
cleaned_lat <- data.frame(cleaned_geolocations %>% filter(latitude>100))
cleaned_lat2 <- data.frame(cleaned_geolocations %>% filter(latitude <= 100))
```

```r
cleaned_lat$latitude <- cleaned_lat$X
all_threat_locations <-data.frame(rbind(cleaned_lat2, cleaned_lat))
all_threat_locations$X <- NULL
```

```r
dim(all_threat_locations)
```

```r
names(honeypot)
```

```r
RES <- data.frame(all_threat_locations %>% dplyr::group_by(servername,country, locale, longitude,latitude) %>% dplyr::summarize(count=n()) %>% arrange(-count))
```

```r
top10 <-data.frame(RES %>% top_n(10))
top10$fullIP <-paste0(top10$servername, '(', top10$locale,')')
```

```r
top10
```

```r
ggplot(top10, aes(x = country, y = count)) + geom_bar(stat="identity", fill = "#C8A2CB") + theme_clean() + coord_flip()
```

```r
ggplot(top10, aes(x = locale, y = count)) + geom_bar(stat="identity", fill = "#C8A2CB") + theme_clean() + coord_flip()
```
